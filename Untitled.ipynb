```python
import random
import tkinter as tk

def countdown(time_left):
    if time_left > 0:
        time_label.config(text=str(time_left))
        window.after(1000, countdown, time_left-1)
    else:
        time_label.config(text="Time's up!")
        game_over()

def game_over():
    guess_button.config(state=tk.DISABLED)
    if num_guesses == max_guesses:
        result_label.config(text="Sorry, you ran out of guesses.")
    elif time_left == 0:
        result_label.config(text="Sorry, you ran out of time.")
    else:
        result_label.config(text=f"Congratulations! You completed level {level}.")
        next_level_button.pack()

def next_level():
    global level
    global min_number
    global max_number
    global number
    global max_guesses
    global num_guesses
    global max_time
    global time_left
    level += 1
    min_number = 1
    max_number += 10 # decrease the range of numbers
    number = random.randint(min_number, max_number)
    instruction_label.config(text=f"Level {level}: I'm thinking of a number between {min_number} and {max_number}.")
    max_guesses -= 1 # decrease the number of guesses
    num_guesses = 0
    max_time -= 5 # decrease the time limit
    time_left = max_time
    result_label.config(text="")
    guess_button.config(state=tk.NORMAL)
    next_level_button.pack_forget()
    countdown(max_time)

def check_guess():
    global num_guesses
    global time_left
    guess = int(guess_entry.get())
    num_guesses += 1
    if guess < number:
        result_label.config(text="Your guess is too low.")
    elif guess > number:
        result_label.config(text="Your guess is too high.")
    else:
        result_label.config(text=f"Congratulations! You completed level {level}.")
        guess_button.config(state=tk.DISABLED)
        next_level_button.pack()
    if num_guesses == max_guesses or time_left == 0:
        game_over()

level = 1
min_number = 1
max_number = 10 + level * 10 # decrease the range of numbers
number = random.randint(min_number, max_number)
max_guesses = 10 - level # decrease the number of guesses
num_guesses = 0
max_time = 60 - level * 5 # decrease the time limit
time_left = max_time

window = tk.Tk()
window.title("Guess the Number")
window.configure(bg="#ADD8E6") # change background color

instruction_label = tk.Label(window, text=f"Level {level}: I'm thinking of a number between {min_number} and {max_number}.", font=("Arial", 18), bg="#ADD8E6")
instruction_label.pack()

guess_label = tk.Label(window, text="Enter your guess:", font=("Arial", 14), bg="#ADD8E6")
guess_label.pack()

guess_entry = tk.Entry(window)
guess_entry.pack()

guess_button = tk.Button(window, text="Guess", command=check_guess)
guess_button.pack()

next_level_button = tk.Button(window, text="Next Level", command=next_level)

result_label = tk.Label(window, font=("Arial", 14), bg="#ADD8E6")
result_label.pack()

time_label = tk.Label(window, font=("Arial", 24), bg="#ADD8E6")
time_label.pack()
countdown(max_time)

window.mainloop()
```
